```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro

def check_brownian_motion(series):
    # Calculate increments
    increments = series.diff().dropna()

    # Test for normality of increments
    stat, p_value = shapiro(increments)
    print("Shapiro-Wilk Test for Normality of Increments")
    print("Statistics=%.3f, p-value=%.3f" % (stat, p_value))

    # A common rule: if p < 0.05, we reject the hypothesis that the dataset has a Gaussian distribution.
    if p_value > 0.05:
        print("Increments look Gaussian (fail to reject H0)")
    else:
        print("Increments do not look Gaussian (reject H0)")

    # Optional: Test for homoscedasticity (constant variance) if you have timestamps or known time intervals

    # Return result of normality test as it is the most fundamental property here
    return p_value > 0.05

# Example usage:
# Create a sample Brownian motion series
np.random.seed(0)
time_steps = times
brownian_motion = pd.Series(coupon)

# Check if it follows Brownian motion
result = check_brownian_motion(brownian_motion)
```

# Black Scholes pricing

```python
mean = 9999.900983333333
std = 88.75266514702373
t = 250
# S = coconut
K = 10000
r = 0.001811
# d1 = (math.log(S/K) + (r + (std**2)/2)*t) / (std * math.sqrt(t))
# d2 = d1 - std * math.sqrt(t)

def N(x):
    # return sp.stats.norm.cdf(x, mean, std**2)
    return 0.5 * (1 + math.erf((x-mean) / (std**2*math.sqrt(2))))

# N = norm.cdf
def BS_CALL(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * N(d1) - K * np.exp(-r*T)* N(d2)

def bs_array(arr):
    return [BS_CALL(S, K, t, r, std) for S in arr]

bs_pricing = bs_array(coconut)
mean_coupon = 635.0464333333333
bs_mean = np.mean(bs_pricing)

diff = bs_pricing - bs_mean

factor = 3.229
predicted_coupon = bs_pricing + diff * factor

plt.figure(figsize=(20, 10))

plt.plot(times, predicted_coupon, label='predicted coupon')

print(f"coupon mean: {mean_coupon}, bs mean: {bs_mean}")

plt.plot(times, np.ones(size) * bs_mean, label='mean BS')
plt.plot(times, np.ones(size) * mean_coupon, label='mean coupon')
plt.plot(times, bs_pricing, label='BS')
plt.plot(times, coupon, label='actual coupon price')
plt.legend()
```

```python
# find factor that minimizes RMSE
ranges = np.arange(3.2, 3.4, 0.0001)

best_rmse = 1e9
factor = -1
for i in ranges:
    predicted_coupon = bs_pricing + diff * i

    rmse = np.sqrt(np.mean((predicted_coupon-coupon)**2))

    if rmse < best_rmse:
        best_rmse = rmse
        factor = i

print(best_rmse, factor)
```

# Implied Volatility

```python
from scipy.stats import norm
from scipy.optimize import newton

def black_scholes_price(S, K, t, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * t) / (sigma * np.sqrt(t))
    d2 = d1 - sigma * np.sqrt(t)
    price = S * norm.cdf(d1) - K * np.exp(-r * t) * norm.cdf(d2)
    return price

def implied_volatility(market_price, S, K, t, r, initial_guess=0.2):
    objective_function = lambda sigma: black_scholes_price(S, K, t, r, sigma) - market_price
    return newton(objective_function, initial_guess)
```

```python
# manual newtons method
import numpy as np

def f(x):
    return x**2 - 3  # Example function

def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

def newtons_method(f, x0=0.02, epsilon=1e-7, max_iter=100, h=1e-5):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        if abs(fx) < epsilon:
            return x
        dfx = numerical_derivative(f, x, h)
        if dfx == 0:
            raise ValueError("Derivative zero. No solution found.")
        x -= fx / dfx
    raise ValueError("Maximum iterations reached. No solution found.")

# Example usage
root = newtons_method(f)
```

```python
# get implied volatility of the options
IV = [implied_volatility(c, coco, 10000, 250, 0) for c, coco in zip(coupon, coconut)]
```

```python
# manual iv
man_iv = [newtons_method(lambda sigma: black_scholes_price(coco, 10000, 250, 0, sigma) - c) for c, coco in zip(coupon, coconut)]
```

# Check if difference in coupon and BS prices is mean reverting

```python
import numpy as np

def hurst_exponent(time_series):
    """Calculate the Hurst Exponent of the time series."""
    # Define the time range
    tau = np.arange(1, len(time_series)//2)
    variances = [np.var(np.subtract(time_series[lag:], time_series[:-lag])) for lag in tau]

    # Regression
    poly = np.polyfit(np.log(tau), np.log(variances), 1)
    hurst = poly[0] / 2

    return hurst

def is_mean_reverting(time_series):
    H = hurst_exponent(time_series)
    print(f"Hurst Exponent: {H}")

    if H < 0.5:
        print("The series is likely mean-reverting.")
        return True
    elif H > 0.5:
        print("The series is trending.")
        return False
    else:
        print("The series is a random walk.")
        return False

# Example usage:
# Generate some example data
# Check if the series is mean-reverting

is_mean_reverting(ratio)

# is_mean_reverting(predicted_coupon-coupon)
# is_mean_reverting(bs_pricing-coupon)
```
