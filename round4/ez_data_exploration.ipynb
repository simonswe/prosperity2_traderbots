```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
paths = ["prices_round_4_day_1.csv", "prices_round_4_day_2.csv", "prices_round_4_day_3.csv"]

df = pd.DataFrame()
for i, path in enumerate(paths):
    new_df = pd.read_csv(path, delimiter=";")
    # add 1000000 to timestamp column
    new_df["timestamp"] = new_df["timestamp"] + 1000000 * i
    df = pd.concat([df, new_df])

df.tail(20)
```

```python
size = 30000
times = [i for i in range(size)]
```

```python
bid_volume = df[df["product"] == "COCONUT"]['bid_volume_1']
ask_volume = df[df["product"] == "COCONUT"]['ask_volume_1']
total_volume = bid_volume + ask_volume
coconut_volume = np.array(total_volume.to_list()[0:size])
normalized_coconut_volume = (coconut_volume - np.mean(coconut_volume)) / np.std(coconut_volume)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(times, normalized_coconut_volume, label='cocunut volume')
```

```python
coconut = np.array(df[df["product"] == "COCONUT"]['mid_price'].to_list()[0:size])
coupon = np.array(df[df["product"] == "COCONUT_COUPON"]['mid_price'].to_list()[0:size])

COCONUT_MEAN = np.mean(coconut)
COUPON_MEAN = np.mean(coupon)
```

```python

ranges = [0.0001 * i + 0.5 for i in range(1000)]
print(ranges)

best_rmse = 1e9
factor = -1
for i in ranges:
    predicted_coupon = [(coco_price - COCONUT_MEAN) * i + COUPON_MEAN for coco_price in coconut]

    rmse = np.sqrt(np.mean((predicted_coupon-coupon)**2))

    if rmse < best_rmse:
        best_rmse = rmse
        factor = i

print(best_rmse, factor)
```

```python
factor = coupon/coconut
mean = np.mean(factor)

running_mean = np.zeros(size)
for i in range(size):
    running_mean[i] = np.mean(factor[:i])


plt.figure(figsize=(20, 10))

plt.plot(times, np.ones(size) * mean, label='mean')
plt.plot(times, running_mean, label='running mean')
plt.plot(times, factor, label='factor')
plt.legend()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro

def check_brownian_motion(series):
    # Calculate increments
    increments = series.diff().dropna()

    # Test for normality of increments
    stat, p_value = shapiro(increments)
    print("Shapiro-Wilk Test for Normality of Increments")
    print("Statistics=%.3f, p-value=%.3f" % (stat, p_value))

    # A common rule: if p < 0.05, we reject the hypothesis that the dataset has a Gaussian distribution.
    if p_value > 0.05:
        print("Increments look Gaussian (fail to reject H0)")
    else:
        print("Increments do not look Gaussian (reject H0)")

    # Optional: Test for homoscedasticity (constant variance) if you have timestamps or known time intervals

    # Return result of normality test as it is the most fundamental property here
    return p_value > 0.05

# Example usage:
# Create a sample Brownian motion series
np.random.seed(0)
time_steps = times
brownian_motion = pd.Series(coupon)

# Check if it follows Brownian motion
result = check_brownian_motion(brownian_motion)
```
