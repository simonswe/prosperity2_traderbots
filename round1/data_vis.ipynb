```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv delimited by semi colon
paths = ["prices_round_1_day_-2.csv", "prices_round_1_day_-1.csv", "prices_round_1_day_0.csv"]
# initialize df
df = pd.DataFrame()
for i, path in enumerate(paths):
    # concat df
    new_df = pd.read_csv(path, delimiter=";")
    # add 1000000 to timestamp column
    new_df["timestamp"] = new_df["timestamp"] + 1000000 * i
    df = pd.concat([df, new_df])
```

```python
# remove product where the value is "AMETHYSTS"
df = df[df["product"] != "AMETHYSTS"]
```
